# K-mean Clustering

```python
import pathlib
import imageio
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.color import rgb2gray

# import images
images_at = pathlib.Path('../random-forest/data/images/raw/segment_series.tif')
images = imageio.volread(images_at)
shp = images.shape
print(shp) # [image num, x, y]
```

```python
image = images[0].reshape((shp[1]*shp[2], 1))
plt.imshow(images[0], cmap='gray') 
```

#### In this exercise, I want to distinguish between the bubbles and background. Hence, I'm using 2 clusters.

```python
from sklearn.cluster import KMeans

n_clusters = 2
clustering = KMeans(n_clusters=n_clusters).fit(image)
image_pred = clustering.predict(image).reshape((shp[0], shp[1]))
```

```python
# Stacking the original image with the result image.
plt.imshow(images[0], cmap='gray')
plt.imshow(image_pred, alpha=0.3, cmap='PRGn')
```

#### Let's try using 3 clusters.

```python
n_clusters = 3
clustering4 = KMeans(n_clusters=n_clusters).fit(image)
image_pred4 = clustering4.predict(image).reshape((shp[0], shp[1]))
# Stacking the original image with the result image.
plt.imshow(images[0], cmap='gray')
plt.imshow(image_pred4, alpha=0.3, cmap='PRGn')
```
